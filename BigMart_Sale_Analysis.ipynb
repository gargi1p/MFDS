```python
import pandas as pd
import numpy as num
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df=pd.read_csv(url)
df.head()
mean_value = df['Fare'].mean()
print("mean_value:",mean_value)
median_value = df['Fare'].median()
print("median_value:",median_value)
mode_value = df['Fare'].mode()
print("mode_value:",mode_value)
var_value = df['Fare'].var()
print("variance_value:",var_value)
sd_value = df['Fare'].std()
print("standard_deviation:",sd_value)
skew_value = df['Fare'].skew()
print("skewness_value:",skew_value)
kur_value = df['Fare'].kurt()
print("kurotosis_value:",kur_value)
df['Fare'].value_counts().plot(kind='pie', title='TITANIC')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()

```
